```python
import torch
from transformers import AutoProcessor, AutoModelForCausalLM
from PIL import Image

# Define the model path (adjust as necessary)
model_dir = "/content/drive/MyDrive/radiology_image_captioning_model1"

# Load processor and model
processor = AutoProcessor.from_pretrained(model_dir)
model = AutoModelForCausalLM.from_pretrained(model_dir)
```

```python
# Load your image
image_path = "/content/drive/MyDrive/radio_validate/ROCOv2_2023_train_000443_168.png"
image = Image.open(image_path).convert("RGB")

# Preprocess the image
inputs = processor(images=image, return_tensors="pt")

# Generate predictions
pixel_values = inputs["pixel_values"]
generated_ids = model.generate(pixel_values=pixel_values, max_length=50, num_beams=5)

# Decode the generated caption
caption = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
print("Generated Caption:", caption)

# Display the image
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.title(caption) # Set title as caption
plt.show()
```
